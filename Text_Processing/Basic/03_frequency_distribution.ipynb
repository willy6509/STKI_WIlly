# Frequency Distribution
Menghitung frekuensi kemunculan setiap tokens(kata) dalam teks.

```python
import nltk
import string

from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

kalimat = "Dalam survei ini, responden yang sangat setuju menyatakan bahwa mereka sangat puas, sementara responden yang hanya setuju menyatakan bahwa kepuasan mereka hanya moderat.."
kalimat = kalimat.translate(str.maketrans('','',string.punctuation)).lower()

tokens = nltk.tokenize.word_tokenize(kalimat)
kemunculan = nltk.FreqDist(tokens)

print(kemunculan.most_common())
```

## Frequency Distribution Visualization with Matplotlib
Menggambarkan frekuensi kemunculan setiap tokens dapat menggunakan library _matplotlib_ pada Python.

https://matplotlib.org

```python
import matplotlib.pyplot as plt

kemunculan.plot(30, cumulative=False)

plt.show()
```
